Creating an transition matrix of Blake Snell's pitch-by-pitch data.
Left to right: fastball, changeup, curveball, slider.

```python
T=np.array([[0.372670807453416, 0.221238938053097, 0.402439024390244, 0.275510204081633], [0.180124223602484, 0.300884955752212, 0.26219512195122, 0.244897959183673], [0.285714285714286, 0.283185840707965, 0.195121951219512, 0.275510204081633], [0.161490683229814, 0.194690265486726, 0.140243902439024, 0.204081632653061]])
print(T)
```

Initial vector of pitch distribution.

```python
v=np.array([0.25,0.25,0.25,0.25])
print(v)
```

Running the markov chain to the 100,000th power to determine the likelihood of a given pitch over the long-haul.

```python
print(np.dot(np.linalg.matrix_power(T,100000),v))
```
